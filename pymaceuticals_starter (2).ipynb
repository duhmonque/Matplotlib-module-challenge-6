## Summary Statistics

```python
#Generate a summary statistics table consisting of the mean, median, variance,standard deviation, and SEM of the tumor volume for each drug regimen.
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 

# mean of the tumor volume for each drug regimen
mean_df = complete_data_results.groupby("Drug Regimen").mean()["Tumor Volume (mm3)"]
mean_df
```

```python
# median of the tumor volume for each drug regimen
median_df = complete_data_results.groupby("Drug Regimen").median()["Tumor Volume (mm3)"]
median_df
```

```python
# variance of the tumor volume for each drug regimen
variance_df = complete_data_results.groupby("Drug Regimen").var()["Tumor Volume (mm3)"]
variance_df
```

```python
# standard deviantion of the tumor volume for each drug regimen
std_df = complete_data_results.groupby("Drug Regimen").std()["Tumor Volume (mm3)"]
std_df
```

```python
##Standard error of mean of the tumor volume for each drug regimen
sem_df = complete_data_results.groupby("Drug Regimen").sem()["Tumor Volume (mm3)"]
sem_df
```

```python
### Assemble the resulting series into a single summary dataframe.
#Merging statistical summary above
summary_df = pd.DataFrame(
    {"Mean Tumor Volume":mean_df,
     "Median Tumor Volume":median_df,
     "Tumor Volume Variance":variance_df,
     "Tumor Volume Std. Dev.":std_df,
     "Tumor Volume Std. Err.":sem_df     
    })
summary_df
```

```python

# Using the aggregation method, produce the same summary statistics in a single line

summary_df = pd.DataFrame(
    {"Mean":mean_df,
     "Median":median_df,
     "Var":variance_df,
     "Std":std_df,
     "SEM":sem_df},   
     columns=['Mean', 'Median', 'Var', 'Std', 'SEM'])
summary_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
Capomulin_df = new_data_frame.loc[new_data_frame["Drug Regimen"] == "Capomulin",:]
Ramicane_df = new_data_frame.loc[new_data_frame["Drug Regimen"] == "Ramicane", :]
Infubinol_df = new_data_frame.loc[new_data_frame["Drug Regimen"] == "Infubinol", :]
Ceftamin_df = new_data_frame.loc[new_data_frame["Drug Regimen"] == "Ceftamin", :]

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
Capomulin_last = Capomulin_df.groupby('Mouse ID').max()['Timepoint']
Capomulin_vol = pd.DataFrame(Capomulin_last)
Capomulin_merge = pd.merge(Capomulin_vol, new_data_frame, on=("Mouse ID","Timepoint"),how="left")
Capomulin_merge
```

```python
# Put treatments into a list for for loop (and later for plot labels)
regimen_list = ["Capomulin", "Ramicane","Infubinol", "Ceftamin"]
regimen_list
# Create empty list to fill with tumor vol data (for plotting)

empty_list = []
#empty_dictionary = {}

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for regimen in regimen_list:
    tumor_volume_subset = complete_data_results.loc[complete_data_results["Drug Regimen"] == regimen,"Tumor Volume (mm3)"]
    empty_list.append(tumor_volume_subset)
    #empty_dictionary[regimen]= list(tumor_volume_subset)
    
    
    #print(tumor_volume_subset)
    # Locate the rows which contain mice on each drug and get the tumor volumes
    
    
    # add subset 
    
    
    # Determine outliers using upper and lower bounds
    quartiles = tumor_volume_subset.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers = tumor_volume_subset[(tumor_volume_subset < lower_bound) | (tumor_volume_subset > upper_bound)]
    
    print(f"{regimen} potential outliers: {outliers}")

# drug_volume = pd.DataFrame(empty_dictionary)
# drug_volume.head()
#empty_list
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
# Example outlier plot of reaction times
fig1, ax1 = plt.subplots()
ax1.set_ylabel('Final Tumor Volume (mm3)')
ax1.boxplot(empty_list, labels = regimen_list)
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

corr=round(st.pearsonr(ave_tumvol2['Weight (g)'],ave_tumvol2['Tumor Volume (mm3)'])[0],2)
print(f"The correlation between mouse weight and average tumor volume is {corr}")







```

```python
x_values = ave_tumvol2['Weight (g)']
y_values = ave_tumvol2['Tumor Volume (mm3)']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept

print(f"slope:{slope}")
print(f"intercept:{intercept}")
print(f"rvalue (Correlation coefficient):{rvalue}")
print(f"pandas (Correlation coefficient):{corr}")
print(f"stderr:{stderr}")

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

print(line_eq)
```

```python

# Add the linear regression equation and line to plot

plt.scatter(x_values,y_values,s=175, color="blue")
plt.plot(x_values,regress_values,"r-")
plt.title('Regression Plot of Mouse Weight Versus Average Tumor Volume',fontsize =20)
plt.xlabel('Weight(g)')
plt.ylabel('Average Tumore Volume (mm3)')
ax1.annotate(line_eq, xy=(20, 40), xycoords='data',xytext=(0.8, 0.95), textcoords='axes fraction',horizontalalignment='right', verticalalignment='top',fontsize=30,color="red")

print(f"The r-squared is: {rvalue**2}")

plt.show()
```
